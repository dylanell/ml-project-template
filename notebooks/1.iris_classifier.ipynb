# Iris Classifier

```python
import json

from ml_project_template.utils import seed_everything

seed = 42
seed_everything(seed)
```

### Model Registry

```python
from ml_project_template.models import ModelRegistry
print(ModelRegistry.list())
```

### Initialize Datasets

```python
from ml_project_template.data import TabularDataset

train_data = TabularDataset(
    csv_path=".data/iris/processed/train.csv",
    target_col="species"
)

val_data = TabularDataset(
    csv_path=".data/iris/processed/valid.csv",
    target_col="species"
)
```

### Train Gradient Boosting Classifier

```python
model = ModelRegistry.get("gb_classifier")(
    n_estimators=10, 
    max_depth=3
)

model.train(
    experiment_name="iris-gb-classifier",
    train_data=train_data,
    val_data=val_data,
    model_path=".models/iris_gb_classifier",
    tracking=True
)
```

### Evaluate Gradient Boosting Classifier

```python
model = ModelRegistry.load(".models/iris_gb_classifier_final", device="cpu")

metrics = model.evaluate(val_data)
print(json.dumps(metrics, indent=2))
```

### Train MLP Classifier

```python
seed_everything(seed)

model = ModelRegistry.get("mlp_classifier")(
    layer_dims=[4, 16, 3],
    hidden_activation="LeakyReLU",
    output_activation="Identity",
    use_bias=True,
    norm="layer"
)

model.train(
    experiment_name="iris-mlp-classifier",
    train_data=train_data,
    val_data=val_data,
    model_path=".models/iris_mlp_classifier",
    tracking=True,
    lr=1e-3,
    weight_decay = 0.01,
    max_epochs=1000,
    batch_size=8,
    val_frequency=1,
    patience=10
)
```

### Evaluate MLP Classifier

```python
model = ModelRegistry.load(".models/iris_mlp_classifier_best")

metrics = model.evaluate(val_data)
print(json.dumps(metrics, indent=2))
```
